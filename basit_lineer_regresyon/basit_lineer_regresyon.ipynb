```python
import pandas as pd
import numpy as np
import seaborn as sbn

df = pd.read_csv("Advertising.csv")
df = df.iloc[:,1:len(df)]
df.head()
```

```python
df.info()
```

```python
sbn.jointplot(x = "TV", y = "sales",data = df,kind = "reg")
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = df[["TV"]]
X.head()
```

```python
y = df[["sales"]]
y.head()
```

```python
model = LinearRegression().fit(X,y)
```

```python
model
```

```python
dir(model)
```

##### intercept --> sabit (b0)
##### coef_  --> b1 değeri

```python
model.intercept_
```

```python
model.coef_
```

```python
#rkare  --> Bağımlı değişkendeki değişikliğin, bağımsız değişkenlerce açıklanma yüzdesidir.
model.score(X,y)
```

```python
gercek_y = y[0:10]
tahmin_y = pd.DataFrame(model.predict(X)[0:10])
```

```python
hatalar = pd.concat([gercek_y,tahmin_y],axis=1)
```

```python
hatalar.columns = ["gercek_y","tahmin_y"]
```

```python
hatalar
```

```python
hatalar["hata"] = hatalar["gercek_y"] - hatalar["tahmin_y"]
```

```python
hatalar
```

```python
## mse
hatalar["hatalar_kare"] = hatalar["hata"]**2
```

```python
hatalar
```

```python
np.mean(hatalar["hatalar_kare"])
```
